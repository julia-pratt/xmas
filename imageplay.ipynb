to wrap your gift i took an image and 
first: rotated all the rows by three 
second: rotated all of the columns by four

I think the index of the row and the column get multiplied by the numbers I give you
And then I shuffle by the result
And you alternate direction on the way down

What he did:
 row index * 3 (+ - downwards)
 column index * 4 (+ - downwards)

What I would need to do:
 column index * 4 (- + downwards)
 row index * 3 (- + downwards)

```python
import numpy as np
from PIL import Image
```

```python
image = Image.open('./wrapped.png')
original = np.array(image)
original.shape
Image.fromarray(original).show()
```

```python
buffer = np.array(image)
```

```python
# the columns
for j in range(original.shape[1]):
    if j ==1:
        print(original[:, j].shape)
    if j % 2:
        num = 4*-j
    else:
        num = 4*j
    buffer[:, j] = np.roll(original[:, j], num)
Image.fromarray(buffer).show()
```

```python
# the rows
for i in range(original.shape[0]):
    if i ==1:
        print(original[i].shape)
    if i % 2:
        num = 3*-i
    else:
        num = 3*i
    buffer[i] = np.roll(original[i], num)
Image.fromarray(buffer).show()
```

```python
Image.fromarray(original).save('unwrapped.png')
```
